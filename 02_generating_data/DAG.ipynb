# Environment

## Importing modules

```r
library(dagitty)
library(ggdag)
library(dplyr)
library(sf)
library(sp)
library(ggplot2)
```

## Creating DAG

```r
dag_assumed <- dagitty('dag {
C1 [pos="-0.596,-1.658"]
C2 [pos="-0.570,-0.721"]
C3 [pos="-0.596,-0.093"]
Cs [pos="-0.544,0.707"]
T [exposure,pos="-2.200,1.597"]
Y [outcome,pos="1.400,1.621"]
C1 -> T
C1 -> Y
C2 -> T
C2 -> Y
C3 -> T
C3 -> Y
Cs -> T
Cs -> Y
T -> Y
}
'
)
```

```r
ggdag(dag_assumed, text = TRUE, use_labels = "name") +
  theme_dag() 
```
